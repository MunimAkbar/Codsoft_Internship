```python
import random

# Rock
rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""

# Paper
paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""

# Scissors
scissors = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""
game_images = [rock, paper, scissors]

user_score = 0
computer_score = 0

while True:
    user_choice = int(input("Enter 0 for rock, 1 for paper, and 2 for scissors: "))
    
    if user_choice >= 3 or user_choice < 0:
        print("You typed an invalid number. You lose this round.")
    else:  
        print(game_images[user_choice])

        computer_choice = random.randint(0, 2)
        print(f"The computer chose: {computer_choice}")
        print(game_images[computer_choice])
        
        if user_choice == 0 and computer_choice == 1:
            print("You lose this round.")
            computer_score += 1
        elif user_choice == 1 and computer_choice == 0:
            print("You win this round.")
            user_score += 1
        elif user_choice == 2 and computer_choice == 0:
            print("You lose this round.")
            computer_score += 1
        elif user_choice == 0 and computer_choice == 2:
            print("You win this round.")
            user_score += 1
        elif user_choice == 1 and computer_choice == 2:
            print("You lose this round.")
            computer_score += 1
        elif user_choice == 2 and computer_choice == 1:
            print("You win this round.")
            user_score += 1
        else:
            print("It's a draw this round.")
        
        print(f"Your Score: {user_score}")
        print(f"Computer Score: {computer_score}")
    
    play_again = input("Do you want to play another round? (yes/no): ")
    if play_again.lower() != "yes":
        break
        
if user_score > computer_score:
    print("Congratulations! You win the game.")
elif user_score < computer_score:
    print("The computer wins the game. Better luck next time.")
else:
    print("It's a tie game.")
print("Thanks for playing!")
```
